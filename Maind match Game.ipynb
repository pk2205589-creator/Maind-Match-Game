```python
import tkinter as tk
from tkinter import messagebox
import random

root = tk.Tk()
root.title("Memory Match Game")
root.geometry("650x700")
root.config(bg="#0F172A")

player_name = ""

emojis = ["🍎","🍌","🐶","🐱","⭐","❤️","🌸","🎈"] * 2
random.shuffle(emojis)

buttons = []
first = None
second = None
lock = False
score = 0
moves = 0
matched = 0


def start_game():
    global player_name

    player_name = name_entry.get().strip()

    if player_name == "":
        messagebox.showwarning(
            "Warning",
            "Please Enter Your Name!"
        )
        return

    start_frame.pack_forget()

    welcome_label.config(
        text=f"🎉 Welcome {player_name}!"
    )

    game_frame.pack(pady=20)


def check_match():
    global first, second, lock
    global score, matched

    if emojis[first] == emojis[second]:

        buttons[first].config(
            bg="#22C55E"
        )

        buttons[second].config(
            bg="#22C55E"
        )

        score += 10
        matched += 1

        if matched == 8:
            messagebox.showinfo(
                "Winner",
                f"🏆 Congratulations {player_name}!\nYou Won!"
            )

    else:
        buttons[first].config(text="❓")
        buttons[second].config(text="❓")

    score_label.config(
        text=f"🏆 Score: {score}"
    )

    first = None
    second = None
    lock = False


def card_click(index):
    global first, second
    global lock, moves

    if lock:
        return

    if buttons[index]["text"] != "❓":
        return

    buttons[index].config(
        text=emojis[index]
    )

    if first is None:
        first = index

    elif second is None:
        second = index

        moves += 1

        moves_label.config(
            text=f"📊 Moves: {moves}"
        )

        lock = True

        root.after(
            800,
            check_match
        )


def reset_game():
    global first, second
    global score, moves
    global matched
    global lock

    random.shuffle(emojis)

    first = None
    second = None
    score = 0
    moves = 0
    matched = 0
    lock = False

    score_label.config(
        text="🏆 Score: 0"
    )

    moves_label.config(
        text="📊 Moves: 0"
    )

    for btn in buttons:
        btn.config(
            text="❓",
            bg="#1E3A5F"
        )


# ---------- START SCREEN ----------

start_frame = tk.Frame(
    root,
    bg="#0F172A"
)
start_frame.pack(expand=True)

title = tk.Label(
    start_frame,
    text="🎮 MEMORY MATCH GAME 🎮",
    font=("Comic Sans MS", 24, "bold"),
    bg="#0F172A",
    fg="gold"
)
title.pack(pady=20)

tk.Label(
    start_frame,
    text="Enter Your Name",
    font=("Arial", 14),
    bg="#0F172A",
    fg="white"
).pack()

name_entry = tk.Entry(
    start_frame,
    font=("Arial", 14),
    width=25
)
name_entry.pack(pady=10)

tk.Button(
    start_frame,
    text="▶ START GAME",
    font=("Arial", 14, "bold"),
    bg="#2563EB",
    fg="white",
    command=start_game
).pack(pady=20)

# ---------- GAME SCREEN ----------

welcome_label = tk.Label(
    root,
    text="",
    font=("Arial", 16, "bold"),
    bg="#0F172A",
    fg="#22C55E"
)

score_label = tk.Label(
    root,
    text="🏆 Score: 0",
    font=("Arial", 14, "bold"),
    bg="#0F172A",
    fg="yellow"
)

moves_label = tk.Label(
    root,
    text="📊 Moves: 0",
    font=("Arial", 14, "bold"),
    bg="#0F172A",
    fg="cyan"
)

welcome_label.pack()
score_label.pack()
moves_label.pack()

game_frame = tk.Frame(
    root,
    bg="#0F172A"
)

for i in range(16):

    btn = tk.Button(
        game_frame,
        text="❓",
        width=5,
        height=2,
        font=("Arial", 20, "bold"),
        bg="#1E3A5F",
        fg="white",
        activebackground="#3B82F6",
        command=lambda i=i: card_click(i)
    )

    btn.grid(
        row=i // 4,
        column=i % 4,
        padx=5,
        pady=5
    )

    buttons.append(btn)

tk.Button(
    root,
    text="🔄 Reset Game",
    font=("Arial", 14, "bold"),
    bg="#F59E0B",
    fg="white",
    command=reset_game
).pack(pady=15)

root.mainloop()
```
